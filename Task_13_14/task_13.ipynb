```python
import pandas as pd
```

```python
purchases = [
    {"item": "apple", "category": "fruit", "price": 1.2, "quantity": 10},
    {"item": "banana", "category": "fruit", "price": 0.5, "quantity": 5},
    {"item": "milk", "category": "dairy", "price": 1.5, "quantity": 2},
    {"item": "bread", "category": "bakery", "price": 2.0, "quantity": 3},
]
```

```python
df = pd.DataFrame(purchases)
df
```

```python
(df.price*df.quantity).sum()
```

```python
def total_revenue(purchases):
    df = pd.DataFrame(purchases)
    return (df.price * df.quantity).sum()
```

```python
total_revenue(purchases)
```

```python
def items_by_category(purchases):
    res = dict()
    for value in purchases:
        item = value.get("item")
        category = value.get("category")
        if res.get(category):
            res[category].append(item)
        else:
            res[category] = [item]
    return res
    
```

```python
items_by_category(purchases)
```

```python
def expensive_purchases(purchases, min_price):
    df = pd.DataFrame(purchases)
    return df.query(f"price >= {min_price}").to_dict(orient='records')
```

```python
expensive_purchases(purchases, 1.0)
```

```python
def average_price_by_category(purchases):
    df = pd.DataFrame(purchases)
    return df[["category", "price"]].groupby("category").mean("price").sort_values("price").to_dict().get('price')
```

```python
average_price_by_category(purchases)
```

```python
def most_frequent_category(purchases):
    df = pd.DataFrame(purchases)
    return df.groupby("category").count().sort_values("item", ascending=False).reset_index()["category"][0]
```

```python
most_frequent_category(purchases)
```
